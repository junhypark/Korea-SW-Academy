### Closoure

```
def t():
    def y():
        return 1
    return y()
```

This code has problem that from out of the function, nobody can access to the y()

But, thanks to closoure, we can use y()

If we have bracket, the return value comes well

*Other case*

```
def t():
    def y():
        return 1
    return y

t() >>> return function address
t()() >>> the return value comes well
```

So if you want to use inner-function (y()), use two brackets or n-brackets

Unless, the function will return the address

### Lambda

Lambda is similar with mutuable

When we call lambda, it stores on the stack, instead not using memory

def "name"("param"): return "result"
              |                 |
lambda     "param":         "result"

```
(lambda x, y: x+y)(2,3) >>> 5
```

But you cannot use in this situation as desired

```
adders = []

Since lambda is anonymous function, in the stack it keep wrapped

In order to solve, we give argument as default

```
adders.append(lambda m, n=n: m=n)

### Asterisk

1. 3*4 - Multiply
2. 3**4 - Square is actually 3x3x3x3 so it takes more time
3. a, *b = 1,2,3,4 - a: 1, b: [2,3,4]
4. print(*a) - if a = [1,2,3,4], unpack a and print like 1, 2, 3, 4
4. add) print(**a) - unpack dict
```
def foo(x, y):
    print(x, y)
d = {'x': 1, 'y':2}
foo(**d)
# foo(x=1, y=2) is the same with foo(**d)
```
5. def fun(*args) - can give an argument like fun(1,2,3,4,5)
6. def fun(**args) - pack entered keyword arguments into dictionary
```
def foo(**args):

```
7. def fun(*, a, b) - after the asterisk, arguments a and b must be keyword argument
8. *a, - *a, = 1,2,3 gives [1,2,3] / *a = 1,2,3 gives error
9. from import * - import everything
