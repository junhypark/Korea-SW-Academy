```python
import tensorflow as tf
```

```python
# 만드는 방식이 6가지다 
# (쓰지 말라고하는 방식)
# 함수형 패러다임을 객체 지향으로 만들어서 적용시키는 것이 best

# 1. Sequential 방식
# 문제가 존재 (multi-input / multi-output이 안된다)
model = tf.keras.Sequential()
    # 안의 layer를 구성하는 방식은 두가지
    # 복잡한 것을 만들때는 tensorflow가 더 좋다
```

```python
model.add(tf.keras.layers.Dense(64, input_shape = (4,)))
# 64 dense layer 추가
```

```python
model.add(tf.keras.layers.Dense(64))
```

```python
model.add(tf.keras.layers.Dense(3))
# 하나씩 추가하는 방식
# 유연하게 레이어를 바꿀 수 있다
```

```python
# 시작과 동시에 만드는 방법
model2 = tf.keras.Sequential([
    tf.keras.layers.Dense(64, input_shape = (4,)),
    # tf.keras.layers.ReLU(),                       첫번째 방법
    # tf.keras.layers.Activation('relu')            두번째 방법
    tf.keras.layers.Dense(64, activation='relu'), # 세번째 방법
    tf.keras.layers.Dense(3, activation='softmax')
])
```

```python
# compile을 해야함
model2.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy())   # loss func을 넣을때는 함수로 넣거나 문자열로 넣는다 (함수로 넣어야 옵션 부여 가능)
# computational graph 형태로 바꿔준다
```

```python
model2.summary()
```

```python
model2.fit(iris.iloc[:,:-1], iris.target)
```

```python
model2.fit(iris.iloc[:,:-1], iris.target, epochs=10)
# loss가 안떨어지므로 문제
# fit안써도 되는데
# 전문가들은 직접 학습 코드를 한다
# 이에 맞게 loss func도 구현을 따로 한다
# @tf.function - 알아서 computational graph로 바꿔주는 데코레이터
```
