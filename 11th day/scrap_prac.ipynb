```python
# network에서 단어를 입력했을 때, 빈번히 새로 fetch되는 파일을 봐보자
# ac 파일이 많이 fetch 될 것이다
url = 'https://ac.search.naver.com/nx/ac'
params = {
    'q': '수지',
    'con': 0,
    'frm': 'nv',
    'ans': 2,
    'r_format': 'json',
    'r_enc':'UTF-8',
    'r_unicode': 0,
    't_koreng': 1,
    'run': 2,
    'rev': 4,
    'q_enc':'UTF-8',
    'st':100
}
# 이 params는 해당 파일의 payload를 봐야한다

resp = request('GET', url, params=params)
resp.status_code, resp.headers['content-type']
```
