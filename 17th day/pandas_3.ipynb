```python
# matplotlib

import matplotlib.pyplot as plt
```

```python
# matplotlib의 경우 state machine으로
# 1. 변수를 저장하지 않음. 가장 가까이 있는 것을 찾는다
# 2. 계층적 디폴트 값이 자동 생성된다
# 그래프를 가상 공간에 만드는데 이게 figure()다
```

```python
plt.figure()
```

```python
plt.figure();
# 메모리에 가상 공간을 만든다
```

```python
# plt.figure(figsize = (10, 5))
# 가상 크기 바꿀필요 없으면 생략가능
# axes 밑에는 figure가 있어야함 없으면 자동 생성
# 이것은 모니터 사이즈로 뜬다
plt.axes([0,.5,1,1])
plt.plot([1,2,3,4])
# 여기서 제일 가까운애를 찾는다는 것을
# 가장 가까운 axes를 찾아서 그 위에 그린다는 것이다
plt.grid(True)
# 격자를 그려줌
plt.axes([.5,.5,.5,1])
plt.plot([2,4,5,4])
# 실제 그래프가 생긴다
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.boxplot()
```

```python
sns.boxplot(data =tips, x='day', y='tip')
```

```python
sns.violinplot(data=tips, y='tip')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.boxplot(data=tips, x='sex', y='tip', hue='day')
# size를 키울려면 matplotlib을 사용하여 figure의 figsize를 늘리면 가능하다
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip')
# 전체 데이터를 하나의 직선으로 대표할때 오차범위 보여준다
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.lineplot(data=tips, x= 'sex', y='tip')
```

```python
# map은 1차 형태로 적용됨
# series가 map 적용한다
tips.sex
```

```python
tips.groupby('sex')[['tip']].mean().transform(lambda x:x+100000)
# 이때 transform은 element로 map을 적용한다
```

```python
# apply?
tips.select_dtypes('float64')
# transform이랑 비슷하지만 조금 다름
```

```python
tips.select_dtypes('float64').applymap(lambda x:x+1)
# applymap은 2차에서 사용함
# 하지만 type에 맞는 애들만 가능함 그리고 문자에는 적용 못함
```
