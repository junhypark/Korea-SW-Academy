```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
gnb = GaussianNB()
X_test, X_train, Y_test, Y_train = train_test_split(iris.iloc[:,:-1], iris.target, stratify=iris.target, test_size=0.4)
```

```python
gnb.fit(X_train, Y_train)
```

```python
sum(gnb.predict(X_test) == Y_test) / len(Y_test)

# == 

gnb.score(X_test, Y_test)
# 수치를 그대로 믿을 수 없음
# 왜냐면 hold out에 따라서 달라질 수 있기 때문이다
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(GaussianNB(), iris.iloc[:,:-1], iris.target, cv=10)
```

```python
from sklearn.model_selection import learning_curve
```

```python
# learning curve 그래프 분석
learning_curve(GaussianNB(), iris.iloc[:,:-1], iris.target, cv=10)
```

```python
train_size, train_score, test_score = learning_curve(GaussianNB(), iris.iloc[:,:-1], iris.target, cv=10)
```

```python
data = load_wine(as_frame=True)
```

```python
wine = data.frame
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(wine.iloc[:,:-1], wine.target)
```

```python
# 1. KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(KNeighborsClassifier(), {'n_neighbors':range(2,21), 'algorithm': ['ball_tree', 'kd_tree', 'brute'], 'weights':['uniform', 'distance']}, cv=5)
```

```python
grid.fit(X_train, Y_train)
```

```python
grid.best_estimator_
```
