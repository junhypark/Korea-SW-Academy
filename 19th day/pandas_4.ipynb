```python
# map = series. function. dictionary로 사용 가능
# applymap = dataframe (데이터 전체), function
# transform = groupby
# apply = 함수밖에 사용 못함 (series에서는), dataframe or series에서 사용 가능, aixs 개념을 사용가능하다

import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.sex.apply(lambda x: 'F' if x=='Female' else 'M')
```

```python
tips.select_dtypes('float64').apply(lambda x:x+1)
```

```python
tips
# 비율을 알고 싶을때?
tips.apply(lambda x: x['total_bill']*x['tip'], axis=1)
# column 간의 상관성을 사용해서 함수 만들 수 있다
tips.select_dtypes('float64').apply(lambda x: sum(x))
# axis = 0 이 default
```

```python
# map은 python과 pandas의 map 두개로 나뉜다
# python의 map은 데이터가 뒤에 arg로 들어간다
# pandas의 map은 앞에 data를 적고 dot notation으로 사용 가능하다
list(map(lambda x,y:x+y, [1,2,3,4],[5,6,7,8]))
```

```python
from functools import reduce
```

```python
reduce(lambda x,y:x+y, [1,2,3,4,5])
# reduce는 하나의 data에서 두개씩 뽑아서 사용
```

```python
temp = rating.drop('Timestamp', axis=1)
temp
de = temp.set_index(['UserID', 'MovieID']).unstack().fillna(0)
```

```python
de
```

```python
de1 = de.T.corr()   # column-wise, pair-wise
```

```python
de1[1].nlargest(5)  # 가장 큰 애 5개 뽑기
de1[1].sort_values(ascending=False)[1:5]
```

```python
def sim(user_id, n=5):
    return de1[user_id].sort_values(ascending=False)[1:n+1]
```

```python
sim(3, 3)
```
