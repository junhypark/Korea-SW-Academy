```python
# Naive Bays
# C = {c = class1, class2, ...}
# D = {document1, document2, ...}
# 우리가 수집한 docs 근데 특정 class에 속한 doc이어야 함
# V = {v1, v2, ...}
# documents에 속한 vocab들을 의미
# N = len(document)
```

```python
# data set
trainingSet = [(1, 'Chinese Beijing Chinese', 'yes'), (2, 'Chinese Chinese Shanghai', 'yes'),(3, 'Chinese Macao', 'yes'),(4, 'Tokyo Japan Chinese', 'no')]
testSet = [(5, 'Chinese Chinese Chinese Tokyo Japan')]
```

```python
from nltk.tokenize import word_tokenize
from nltk.text import Text

C = ('yes', 'no')

def training(C, D):
    # 1 line
    V = list()
    for d in D:
        for v in word_tokenize(d[1].lower()):
            if v not in V:
                V.append(v)
    # 2 line
    N = len(D)
    # 3 line
    prior = dict()  # 5
    condprob = dict()   # 10
    for c in C:
        # 4 line
        Nc = len([d for d in D if d[-1] == c])
        # 5 line
        prior[c] = Nc/N
        # 6 line
        textc = Text(word_tokenize('\n'.join([d[1] for d in D if d[-1] == c]).lower()))
        # 7 line
        for v in V:
            # 8 line - 6 line
            Tct = textc.count(v)
            # 9 line
            for v in V:
                # 10 line - smoothing
                if v not in condprob:
                    condprob[v] = dict()

                condprob[v][c] = ((textc.count(v)+1)/(len(V)+textc.vocab().N()))

    return V, prior, condprob
V, prior, condprob = training(C, trainingSet)
```

```python
from math import log

def testing(C, V, prior, condprob, d):
    W = list()
    score = dict()
    for t in d:
        W.append(word_tokenize(t[1].lower()))   # 학습과 동일한 방식으로 토큰화
    for c in C:
        score[c] = log(prior[c])
        for w in W:
            for t in w:
                if t in V:
                    score[c] += log(condprob[t][c])
                # else:
                #     score[c] = log(condprob[t][c])
    return score, max(score, key=score.get)

testing(C, V, prior, condprob, testSet)
```

```python
from nltk.tokenize import word_tokenize
from nltk.text import Text

from konlpy.tag import Okt, Kkma, Mecab, Komoran, Hannanum
from konlpy.data import textwrap

C = ('spam', 'ham')
tokenizer = [Okt(), Kkma(), Komoran(), Hannanum()]

def training(C, D):
    # 1 line
    result_v = dict()
    result_condprob = dict()
    result_prior = dict()

    count = 0

    for token in tokenizer:
        V = list()
        for d in D:
            for v in token.morphs(d[0]):
                if v not in V:
                    V.append(v)
        # 2 line
        N = len(D)
        # 3 line
        prior = dict()  # 5
        condprob = dict()   # 10
        for c in C:
            # 4 line
            Nc = len([d for d in D if d[-1] == c])
            # 5 line
            prior[c] = Nc/N
            # 6 line
            textc = Text(token.morphs('\n'.join([d[0] for d in D if d[-1] == c])))
            # 7 line
            for v in V:
                # 8 line - 6 line
                Tct = textc.count(v)
                # 9 line
                for v in V:
                    # 10 line - smoothing
                    if v not in condprob:
                        condprob[v] = dict()
                    condprob[v][c] = ((textc.count(v)+1)/(len(V)+textc.vocab().N()))
        result_v[count] = V
        result_condprob[count] = condprob
        result_prior[count] = prior
        count += 1

    return result_v, result_prior, result_condprob
V, prior, condprob = training(C, dataset)
```
